```python
# Import the file comparison library
# Thanks https://stackoverflow.com/questions/1072569/see-if-two-files-have-the-same-content-in-python
import filecmp

# Import necessary tools to search filenames
import os

# Import a sorting algorithm that does the legwork lol
# thanks https://stackoverflow.com/questions/33159106/sort-filenames-in-directory-in-ascending-order/33159707
!pip install natsort
import natsort as ns
```

```python
# Defines the 4 lists to be used for comparison (in same cell because it resets)
HR_old = []
HR_new = []
MAP_old = []
MAP_new = []

# Searches files appropriate for each file and adds if relevant
# https://stackoverflow.com/questions/9623398/text-files-in-a-dir-and-store-the-file-names-in-a-list-python
directory = os.listdir("data")
for item in directory:
    if ".dat" in item: # eliminates .m and .csv files
        if "HR" in item:
            if len(item) > 10: # new names have max of 9
                HR_old.append(os.path.abspath("data/{}".format(item)))
            else:
                HR_new.append(os.path.abspath("data/{}".format(item)))
        else: # "MAP"
            if len(item) > 11: # new names have max of 10
                MAP_old.append(os.path.abspath("data/{}".format(item)))
            else:
                MAP_new.append(os.path.abspath("data/{}".format(item)))

# For debug: print all
# print(HR_old, "\n~~~~~~\n", HR_new, "\n~~~~~~\n", MAP_old, "\n~~~~~~\n", MAP_new)
```

```python
# Checks length of lists
print("LIST LENGTHS:  HR_old:", len(HR_old), " HR_new:", len(HR_new),
      " MAP_old:", len(MAP_old), " MAP_new:", len(MAP_new))

# Sorts lists
HR_old = ns.natsorted(HR_old)
HR_new = ns.natsorted(HR_new)
MAP_old = ns.natsorted(MAP_old)
MAP_new = ns.natsorted(MAP_new)

# For debug: print all
# print(HR_old, "\n~~~~~~\n", HR_new, "\n~~~~~~\n", MAP_old, "\n~~~~~~\n", MAP_new)
```

```python
# Temporarily sets working directory to data
# thanks https://stackoverflow.com/questions/1810743/how-to-set-the-current-working-directory

falseCount = 0

# Compares HR files
for i in range (0, 58): # confusingly, index 0 is HR_1 and so forth
    print("\nComparing HR#{}: ".format(i+1), HR_old[i], HR_new[i])
    bool = filecmp.cmp(HR_old[i], HR_new[i])
    print("Result: ", bool)
    if not bool:
        falseCount = falseCount + 1

# Compares MAP files
for i in range (0, 58):
    print("\nComparing MAP#{}: ".format(i+1), MAP_old[i], MAP_new[i])
    bool = filecmp.cmp(MAP_old[i], MAP_new[i])
    print("Result: ", bool)
    if not bool:
        falseCount = falseCount + 1

print("\nNumber of false: ", falseCount)
```
